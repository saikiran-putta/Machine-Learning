```python
from wordcloud import WordCloud, STOPWORDS
```

```python
import PIL.Image
```

```python
text = open('python_intro.txt', 'r').read()
text
```

```python
plt.figure(figsize = (8, 8))
wc = WordCloud(background_color = 'lightgrey', colormap = 'rainbow').generate(text)

plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
pic_python = np.array(PIL.Image.open('python.png'))
```

```python
plt.figure(figsize = (8, 8))
wc = WordCloud(background_color = 'white', colormap = 'winter', mask = pic_python).generate(text)
```

```python
plt.imshow(wc)
plt.axis('off')
plt.show()
```
