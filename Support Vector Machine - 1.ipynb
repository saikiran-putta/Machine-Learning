```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
iris.data
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target == 1]
```

```python
df[df.target == 2]
```

```python
df['flower_name'] = df.target.apply(lambda x : iris.target_names[x])
df.head()
```

```python
df0 = df[:50]
df1 = df[50:100]
df2 = df[100:]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.scatter(df0['sepal length (cm)'], df0['sepal width (cm)'], color = 'red', marker = '+')
plt.scatter(df1['sepal length (cm)'], df1['sepal width (cm)'], color = 'blue', marker = '*')
```

```python
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.scatter(df0['petal length (cm)'], df0['petal width (cm)'], color = 'green', marker = '+')
plt.scatter(df2['petal length (cm)'], df2['petal width (cm)'], color = 'black', marker = '*')
```

```python
X = df.drop(['target', 'flower_name'], axis = 1)
y = df.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict([[4.8, 3.0, 1.7, 0.6]])
```

```python
#Regularization
model_c = SVC(C = 10)
model_c.fit(X_test, y_test)
model_c.score(X_test, y_test)
```

```python
#gamma
model_g = SVC(gamma = 10)
model_g.fit(X_test, y_test)
model_g.score(X_test, y_test)
```

```python
#kernel
model_l_k = SVC(kernel = 'linear')
model_l_k.fit(X_test, y_test)
model_l_k.score(X_test, y_test)
```
